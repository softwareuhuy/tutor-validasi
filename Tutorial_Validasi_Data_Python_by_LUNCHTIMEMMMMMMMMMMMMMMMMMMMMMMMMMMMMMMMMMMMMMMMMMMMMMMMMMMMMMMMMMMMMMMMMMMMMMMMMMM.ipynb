```
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')

# baca path gdrive
# SHEET YANG DIDOWNLOAD BERFORMAT 00:00 GAK USAH GMT+7

# terus ini nama file yang kita coba buat latihan
df=pd.read_csv('/content/gdrive/My Drive/Reyhan/Deli/25_FEB-09_OKT.csv')

df = df.loc[:,['created_at','entry_id','field1','field2','field3','field4','field5','field6','field7']]

# HILANGIN TULISAN T DAN Z YANG ADA
df['created_at']=df['created_at'].str.replace("T"," ")
df['created_at']=df['created_at'].str.replace("Z"," ")

df.rename(columns={'field1': 'Temperatur', 'field2': 'Kelembapan Relatif','field3': 'PM2.5', 'field4' : 'CO2', 'field5' : 'Tekanan Udara','field6' : 'Temperatur SHT31','field7' : 'Humidity SHT31'}, inplace=True)
df = df.dropna()
# UBAH time yang ada jadi gmt +7 local time
df['created_at'] = pd.to_datetime(df['created_at']).dt.tz_convert('Asia/Jakarta')

# hilangkan nilai co2 dan pm2.5 yang bernilai <= 0
df=df[(df['CO2']>0)]
df=df[(df['PM2.5']>0)]


df
```

```
# Ambil Kelembapan Relatif & PM 2.5 (KITA TAMBAHKAN INDEX BULAN, HARI DAN JAM) UNTUK DISORTING NANTI
df1 = df.loc[:,['created_at','entry_id','Temperatur','Kelembapan Relatif','PM2.5','CO2']]
df1['Bulan'] = pd.DatetimeIndex(df['created_at']).month
df1.set_index = df1['Bulan'].inplace=True
df1['Hari'] = pd.DatetimeIndex(df['created_at']).day
df1.set_index = df1['Hari'].inplace=True
df1['Jam'] = pd.DatetimeIndex(df['created_at']).hour
df1.set_index = df1['Jam'].inplace=True

df1
```

```
#Validasi


# terus kita filter data nya per 2 menit(sesuai ama waktu register data thingspeak)
dfco2 = df1.groupby(pd.Grouper(key='created_at',freq='2min')).mean().reset_index()

#Hapus nilai CO2 yang di luar batas range
dfco2 = dfco2[(dfco2['CO2']>400) &  (dfco2['CO2'] <5000)]

# kita ambil nilai index nya sesuai yang ada di petik '',
dfco2 = dfco2.loc[:,['created_at','Temperatur','Kelembapan Relatif','CO2','Bulan','Hari']]

# dropna() fungsinya buat drop data yang kosong
dfco2 = dfco2.dropna()

# astype int ngubah angka ke int
dfco2['CO2']=dfco2['CO2'].round().astype(int)

#round(2) gunanya buat ngatur 2 angka dibelakang koma, bisa atur angka apa aja
dfco2 = dfco2.round(2)
dfco2
```

```
#Ambil data CO2 berdasarkan bulan

# ini path drivenya, kalau mau bikin sendiri, buat dulu foldernya di gdrive terus ganti aja, kalau mau oprek
path = '/content/gdrive/My Drive/Reyhan/Training/Bulan/'

# misal kalian mau ambil bulan ke 5 aja, jadi x=1 diubah jadi x=5
# jadi x ini sebagai inisialisasi bulan awal yang mau dibaca aja
x=1

# disini make while loop tergantung dari kebutuhan, misal kalian cuma mau ambil bulan 5 sampe 12 jadi boolean nya x<=12
# atau kalau mau ambil bulan 5 nya aja, tinggal samain aja variabel x nya sama angkanya (x==5)
while x<=12 :

  # disini dataframenya ambil yang index bulannya sesuai sama variabel x
  dfvalidasi = dfco2[(dfco2['Bulan']==x)]
  # terus dia cuma ambil kolom created at, kelembapan relatif dan co2
  dfvalidasi = dfvalidasi.loc[:,['created_at','Kelembapan Relatif','CO2']]

  # disini gw declare data itu buat baca ada berapa data di dfvalidasi pas bulan tersebut, len(dfvalidasi.axes[0]) sama fungsinya kek count kalo di excel
  data = len(dfvalidasi.axes[0])
  jumlah = str(data)
  # terus di filter, kalau datanya kosong diloncatin pake fungsi pass, jadi langsung lanjut iterasi selanjutnya
  if (dfvalidasi.empty==True):
    pass
  # kalau ada datanya kita lanjut cek outlier pake metode boxplot
  else :
  # buat ini kalian baca sendiri boxplot outlier, karena gw ngajarin python pandas, bukan ngajar probstat
    Q3 = dfvalidasi['CO2'].describe()['75%']
    Q1 = dfvalidasi['CO2'].describe()['25%']
    IQR = Q3 - Q1
    UF = Q3 + (1.5*IQR)
    LF = Q1 - (1.5*IQR)

    # kalau UF sama sama LF udah ketemu di bulannya, jadi data yang diluar range tadi, di drop
    dfvalidasi = dfvalidasi[(dfvalidasi['CO2'] < UF) & (dfvalidasi['CO2'] > LF)]

    # terus kalau udah drop outliernya, tiap 24 jam nya sini disini di rata ratain (kalau pandas mulai dari 00.00 sampe 23.00)
    dfvalidasi = dfvalidasi.groupby(pd.Grouper(key='created_at',freq='24H')).mean().reset_index()

    # disini kita pake argument std buat cari standar deviasi, terus round 2 buat ambil 2 angka dibelakang koma
    std = dfvalidasi['CO2'].std(numeric_only=True).round(2)

    # disini kita ambil rata rata juga
    mean = dfvalidasi['CO2'].mean()

    # cari koefisien variansi
    cv = ((std / mean)*100).round(2)

    # disini gw ubah data yang didapet dijadiin string, biar dia bisa jadi nama buat filenya
    sd = str(std)
    koef = str(cv)
    bul = str(x)

    # kalau gak inget skill issue
    dfvalidasi = dfvalidasi.round(2)

    # terus data yang ke validasi langsung ke simpen sesuai sama nama variabel dan data yang ke itung,
    dfvalidasi.to_csv(path+"Bulan "+bul+" | "+"SD: "+sd+" ppm "+ "CV: " +koef+"% | "+"CO2 Validation.csv", index=None)

  x+=1
```

```
#Ambil data CO2 tiap 1 Harinya
path = '/content/gdrive/My Drive/Reyhan/Training/Hari/'

# Disini sama aja yak caranya, cuma yang beda doang ntar dijelasin, kalau lupa, tinggal scroll atas aja

x=1
while x<=12 :

  # dari sini y=1 (buat hari) itu biar setiap kali ganti bulan, iterasi y nya reset ke 1, jadinya, pas tanggal dah sampe 31, baru x nya ketambah 1, ngulang lagi, terus y nya reset 1 lagi
  y=1
  while y<=31:

    # disini basicnya sama cuma kita tambahin filter hari, jadi buat mastiin kita ambil di hari itu aja di bulannya
    dfvalidasi = dfco2[(dfco2['Bulan']==x) & (dfco2['Hari']==y) ]
    dfvalidasi = dfvalidasi.loc[:,['created_at','Kelembapan Relatif','CO2']]

    data = len(dfvalidasi.axes[0])
    if (dfvalidasi.empty==True):
      pass
    else :
      Q3 = dfvalidasi['CO2'].describe()['75%']
      Q1 = dfvalidasi['CO2'].describe()['25%']
      IQR = Q3 - Q1
      UF = Q3 + (1.5*IQR)
      LF = Q1 - (1.5*IQR)
      dfvalidasi = dfvalidasi[(dfco2['CO2'] < UF) & (dfco2['CO2'] > LF)]

      # kalau tadi kita rata ratain per bulan isi data 24H, kalau sekarang per hari di rata ratain per jam  jadi freq nya 1H
      dfvalidasi = dfvalidasi.groupby(pd.Grouper(key='created_at',freq='1H')).mean().reset_index()
      df5 = dfvalidasi.loc[:,['CO2']]
      std = df5['CO2'].std(numeric_only=True).round(2)
      mean = df5['CO2'].mean()
      cv = ((std / mean)*100).round(2)

      bul = str(x)
      har = str(y)
      sd = str(std)
      koef = str(cv)

      dfvalidasi = dfvalidasi.round(2)
      dfvalidasi.to_csv(path+"Bulan "+bul+" Hari "+har+" | "+"SD: "+sd+" ppm "+ "CV: " +koef+"% | "+"CO2 Validation.csv", index=None)
    y+=1
  x+=1
```

Nah, disini tuh seharusnya kita validasi langsung aja per jam dulu cari outlier, terus kalau data per jamnya kurang dari 22.5 (75% perjam) di drop di jam itu, kalau udah, baru kita sorting.

Kalau kalian bisa, bikin yang kek gitu (yang perjam dulu cari outlier sama kelengkapan data), kalau bisa syukur mantep, tapi kalau gak bisa, penting ngerti aja alur yang gw maksud diatas.
